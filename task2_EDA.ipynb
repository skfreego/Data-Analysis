```python
# 10 Board Feature
list10=[]
for i in data['10board']:
    if ('icse' in i):
        list10.append('ICSE')
    elif('cbse' in i):
        list10.append('CBSE')
    elif('central board of seconary education' in i):
        list10.append('CBSE')
    elif('delhi' in i):
        list10.append('CBSE')
    else:
        list10.append('State Board')
    

data['10 Board'] = list10
data['10 Board'].value_counts()
```
